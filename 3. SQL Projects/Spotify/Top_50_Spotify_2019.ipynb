```python
import pandas as pd 
import sqlite3
```

```python
spotify_2019 = pd.read_csv("top50_spotify_2019.csv", encoding='latin-1')
```

```python
spotify_2019
```

```python
connection = sqlite3.connect('top_50_spotify_2019.db')
```

```python
spotify_2019.to_sql('spotify_2019', connection, if_exists='replace', index=False)
```

```python
query = 'SELECT * FROM  spotify_2019'
```

```python
df = pd.read_sql_query(query, connection)
df
```

```python
print("Number of songs each artist has in Top 50")
query = 'SELECT "Artist.Name", count("Artist.Name") as "Number of Songs in Top 50" from spotify_2019 group by "Artist.Name"'
df = pd.read_sql_query(query, connection)
df
```

```python
print("Most popular song of Top 50 is: ")
query = 'SELECT "Track.Name", "Artist.Name", max(Popularity) as "Popularity" from spotify_2019'
df = pd.read_sql_query(query, connection)
df
```

```python
print("Top 50 ranked")
query = 'SELECT * FROM  spotify_2019 order by Popularity desc'
df = pd.read_sql_query(query, connection)
df
```
