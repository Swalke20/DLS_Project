What might they want to know?  How salaries have increased over time?  What department costs the most?  Predict future salaries to know how to account for salary cost/ budget?

```python
import pandas as pd
import sqlite3
```

```python
city_payroll = pd.read_csv("city_payroll.csv") 
```

```python
city_payroll.info()
```

```python
city_payroll['Total Payments'] = city_payroll['Total Payments'].replace('[\$,]', '', regex=True).astype(float)
city_payroll['Total Payments']
```

```python
connection = sqlite3.connect('city_payroll.db')
cursor = connection.cursor()
```

```python
city_payroll.to_sql('city_payroll', connection, if_exists='replace', index=False)
```

```python
query = 'SELECT * FROM  city_payroll'
df = pd.read_sql_query(query, connection)
df
```

```python
df.info()
```

```python
query = 'SELECT "Year", sum("Total Payments") as "Total Salary Cost" from city_payroll group by "Year"'
df = pd.read_sql_query(query, connection)
df
```

```python
print("Average salary per year across departments")
query = 'SELECT "Year", avg("Total Payments") as "Average Salary Cost" from city_payroll group by "Year"'
df = pd.read_sql_query(query, connection)
df
```

```python
query = 'SELECT distinct "Department Title" FROM  city_payroll'
df = pd.read_sql_query(query, connection)
df
```
