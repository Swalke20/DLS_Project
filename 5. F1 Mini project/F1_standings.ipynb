```python
results.rename(columns={'positionOrder': 'new_points'}, inplace=True)
results
```

```python
results = pd.merge(results, race_years, on='raceId', how='inner')
results = pd.merge(results, drivers_name, on='driverId', how='inner')
```

```python
results
```

```python
new_results = results[['year', 'driverRef', 'points', 'new_points']]
```

```python
summed_df = new_results.groupby(['year', 'driverRef']).agg({'points': 'sum', 'new_points': 'sum'}).reset_index()
```

```python
summed_df_sorted = summed_df.sort_values(by='year')

# Then, sort by 'points' within each 'year' group in descending order
summed_df_sorted_points = summed_df_sorted.groupby('year', group_keys=False).apply(lambda x: x.sort_values(by='points', ascending=False))

# Then, sort by 'points' within each 'year' group in descending order
summed_df_sorted_new_points = summed_df_sorted.groupby('year', group_keys=False).apply(lambda x: x.sort_values(by='new_points', ascending=True))
```
