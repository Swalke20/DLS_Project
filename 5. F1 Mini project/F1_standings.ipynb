**What is the effect of the points system on F1 driver's championship?**  

Hypothesis: To encourage more competitive driving the points system favours the person who comes first, but is someone who comes first and then fourth really better than a driver who comes second twice?

```python
import pandas as pd
```

```python
driver_standings = pd.read_csv("driver_standings.csv")
drivers = pd.read_csv("drivers.csv")
races = pd.read_csv("races.csv")
results = pd.read_csv("results.csv")
status = pd.read_csv("status.csv")
```

```python
race_years = races[['raceId', 'year']]
```

```python
drivers_name = drivers[['driverId', 'driverRef']]
```

```python
results = pd.merge(results, race_years, on='raceId', how='inner')
results = pd.merge(results, drivers_name, on='driverId', how='inner')
```

```python
results
```

```python
new_results = results[['year', 'driverRef', 'points', 'new_points']]
```

```python
summed_df = new_results.groupby(['year', 'driverRef']).agg({'points': 'sum', 'new_points': 'sum'}).reset_index()
```

```python
summed_df_sorted = summed_df.sort_values(by='year')

# Then, sort by 'points' within each 'year' group in descending order
summed_df_sorted_points = summed_df_sorted.groupby('year', group_keys=False).apply(lambda x: x.sort_values(by='points', ascending=False))

# Then, sort by 'points' within each 'year' group in descending order
summed_df_sorted_new_points = summed_df_sorted.groupby('year', group_keys=False).apply(lambda x: x.sort_values(by='new_points', ascending=False))
```

```python
summed_df_sorted_new_points
```

```python
summed_df_sorted_points
```

```python
summed_df_sorted_points = summed_df_sorted_points.sort_values(by=['year', 'points'], ascending=[True, False])
summed_df_sorted_points['Position'] = summed_df_sorted_points.groupby('year')['points'].rank(method='first', ascending=False)
summed_df_sorted_points['Position'] = summed_df_sorted_points['Position'].astype(int)
```

```python
winner_points = summed_df_sorted_points.drop_duplicates(subset=['year'], keep='first')
```

```python
winner_points
```

```python
summed_df_sorted_new_points
```

```python
summed_df_sorted_new_points = summed_df_sorted_new_points.sort_values(by=['year', 'new_points'], ascending=[True, False])
summed_df_sorted_new_points['Position'] = summed_df_sorted_new_points.groupby('year')['new_points'].rank(method='first', ascending=False)
summed_df_sorted_new_points['Position'] = summed_df_sorted_new_points['Position'].astype(int)
```

```python
winner_new_points = summed_df_sorted_new_points.drop_duplicates(subset=['year'], keep='first')
```

```python
winner_new_points
```

```python
drivers_points = winner_points[['year','driverRef']].rename(columns={'driverRef':'Actual_Winner'})
drivers_new_points = winner_new_points[['year', 'driverRef']].rename(columns={'driverRef':'New_points_Winner'})
```

```python
merged_df = pd.merge(drivers_points, drivers_new_points, on='year', how='inner')
```

```python
merged_df['Comparison'] = merged_df['Actual_Winner'] == merged_df['New_points_Winner']
```
