Large loop to go through each row of data.  Each list is either going to be a column in the cleaned data or be used to create a column.  E.g. running total of runs to be used to calculate the DL par score

```python
innings = [0,1]
no_result = 0
dl = 0
match_df = pd.DataFrame()
for index, row in matches.iterrows():
    batting_team = []
    bowling_team = []
    innings_num = []
    over_col = []
    ball_col = []
    match = []
    date = []
    runs = []
    running_total = []
    wickets_taken = []
    batter_out = []
    start_team = []
    remaining_team = []
    venue = []
    winner = []
    missing_result = []
    first_team_innings = []
    first_team_overandball = []
    remaining_over_col = []
    remaining_ball_col = []
    extras=[]
    remainder=[]
    total_overs = []



    try:
        if row['info.outcome']['result']=='no result': 
            no_result+=1
            continue
        if row['info.outcome']['method'] == "D/L":
            dl+=1
            continue

    except KeyError:
        for num in innings:
            cumulative_runs = 0
            batting = row['innings'][num]['team']
            if num == 0:
                bowling = row['innings'][num+1]['team']
            else:
                bowling = row['innings'][num-1]['team']
            starting_team = row['info.players'][batting]
            current_team = starting_team.copy()
            wicket_count = 0


            for over in row['innings'][num]['overs']:
                ball_count = 0
                for ball in row['innings'][num]['overs'][over['over']]['deliveries']:

                    #extras only important because if the ball is a noball or wide then the count doesn't go up - so in an over with a noball there will be 7 deliveries instead of 6
                    if 'extras' in ball:
                        if 'noballs' in ball['extras']:
                            extras.append("noballs")                            

                        elif 'wides' in ball['extras']:
                            extras.append("wides")

                        else:
                            ball_count+=1
                            extras.append("other")
                    else:
                        ball_count+=1
                        extras.append("N/E")
               

                    try:
                        batter_out.append(ball['wickets'][0]['player_out'])
                        if ball['wickets'][0]['kind'] != "retired hurt":
                            wicket_count+=1
                        current_team.remove(ball['wickets'][0]['player_out'])
                        #remove the player from the list current_team
                    except KeyError:
                        #no wicket
                        batter_out.append('NW')
                        pass

                    try:
                        winner.append(row['info.outcome']['winner']) 
                    except:
                        winner.append(row['info.outcome']['result'])


                    batting_team.append(batting)
                    bowling_team.append(bowling)
                    innings_num.append(str(num+1))
                    over_col.append(over['over'])
                    match.append(row['Match ID'])
                    date.append(row['Start Date'])
                    ball_col.append(ball_count)
                    runs.append(ball['runs']['total'])
                    cumulative_runs+=ball['runs']['total']
                    running_total.append(cumulative_runs)
                    wickets_taken.append(wicket_count)
                    start_team.append(starting_team)
                    remaining_team.append(current_team.copy())                    
                    venue_list = [x.strip() for x in row['info.venue'].split(',')]
                    venue.append(venue_list)
                    remaining_over = 49-(over['over'])
                    remaining_ball = 6-ball_count
                    # There are 21 rows where ball count gets to 7 because there's 7 balls listed in the over in the original data.  
                    #No explanation as to why so assumption is that this is an error.  According to search this could be umpire error or manual error inputting data
                    
                    if remaining_ball<0:
                        remaining_ball = 0
                    remaining_over_col.append(remaining_over)
                    remaining_ball_col.append(remaining_ball)

                    #happens when there's a wide or noball on 1st ball of over.
                    if remaining_ball ==6:
                        remainder.append(str(remaining_over +1))

                    #because the dl_df table doesn't have the remainder as 49.0 for example, it has it as 49
                    elif remaining_ball == 0:
                        remainder.append(str(remaining_over))
                    else:
                        remainder.append(str(remaining_over)+"."+str(remaining_ball)) 

            for over in row['innings'][num]['overs']:
                for ball in row['innings'][num]['overs'][over['over']]['deliveries']:
                    if num == 0:
                        first_team_runs = running_total[-1]
                        first_team_innings.append("NaN")
                        first_team_overs = over_col[-1]
                        first_team_balls = ball_col[-1]
                        first_team_overandball.append("NaN")
                    else:
                        first_team_innings.append(first_team_runs)
                        first_team_overandball.append(first_team_overs+first_team_balls)



    loop_df = pd.DataFrame({'batting_team': batting_team, 'bowling_team': bowling_team, 'innings_num': innings_num, 'over_col': over_col, 'ball_col': ball_col, 'Match ID': match, 'Start Date': date, 'Runs': runs, 'Running Total': running_total, 'Batter out': batter_out, 'Wickets taken': wickets_taken, 'Start Team': start_team, 'Remaining Team': remaining_team, 'Venue': venue, 'Winner': winner, 'First Team Innings': first_team_innings, 'Remaining Overs': remaining_over_col, 'Remaining Balls': remaining_ball_col, 'Remainder': remainder, 'Extras': extras, 'Over and Balls Total': first_team_overandball})
    match_df = pd.concat([match_df,loop_df])
```

Iterates through the list of players from the main data, compares the name to a name in the list of choices (a list of names from cricketers_df i.e. ESPN and Wiki) and chooses the closest name to that list.  Then uses *that* name as the key for the dictionary including the values.  Creates a new dictionary called final_cricketers_dict

```python
final_cricketers_dict = {}
for x in player_set:
    name = process.extractOne(x, choices)
    final_cricketers_dict[x] = cricketers_dict[name[0]]

final_cricketers_dict
```

```python
list_team_value = []
average_list_team_value = []
start_list_team_value = []
for index, row in match_df.iterrows():
    team_value = 0
    for player in row['Remaining Team']:
        team_value+=final_cricketers_dict[player]
    average_team_value = team_value/len(row['Remaining Team'])
    start_team_value = 0
    for player in row['Start Team']:
        start_team_value+=final_cricketers_dict[player]
    list_team_value.append(team_value)
    average_list_team_value.append(average_team_value)
    start_list_team_value.append(start_team_value)
```

Calculating the target score to win you round down and add 1.  See the methodology PDF example 4.  But for these purposes I want the score to tie so that it shows if under DL the team would win lose or tie.

```python
dl_percentage_list = []
tie_list = []
count = 0
for delivery_index, delivery_row in match_df.iterrows():
    remaining_overs = delivery_row['Remainder']
    wickets_lost = str(delivery_row['Wickets taken'])
    row_index = np.where(dl_df['Overs Left'] == remaining_overs)[0]
    row_index = row_index[0]
    tie_percentage = dl_df.loc[row_index][wickets_lost]
    tie_percentage = 100-float(tie_percentage)
    dl_percentage_list.append(tie_percentage)
    #int rounds down by default
    tie_score = int(tie_percentage/100*delivery_row['First Team Innings'])
    tie_list.append(tie_score)

dl_percentage_df = pd.DataFrame(dl_percentage_list, columns=['DL %'])
tie_df = pd.DataFrame(tie_list, columns=['Tie Score'])
```

```python
match_df['DL %'] = dl_percentage_df
match_df['Tie Score'] = tie_df

match_df
```

Create a dictionary with the Ground names and RPO, plus the alternative names for the grounds and RPO

```python
#TRIAL
grounds_dict = {}
for index, row in full_grounds_df.iterrows():
    ground = row['Ground']
    grounds_dict[ground] = {"rpo": row['RPO'], "rpw": row['RPW']}
    if row['Known As'] != "NaN": 
        aka = row['Known As']
        grounds_dict[aka] = {"rpo": row['RPO'], "rpw": row['RPW']}

grounds_dict
```

Creating a column for whether the batting team would win under DL

```python
dl_result = []
for index, row in match_df.iterrows():
    if row['Tie Score']<row['Running Total']:
        #batting team wins
        dl_result.append(row['batting_team'])
    elif row['Tie Score']>row['Running Total']:
        #bowling team wins
        dl_result.append(row['bowling_team'])
    elif row['Tie Score']==row['Running Total']:
        #draw
        dl_result.append("Draw")

dl_result_df = pd.DataFrame(dl_result)
match_df['Batting Team Result DL'] = dl_result_df
match_df
```

```python
DL_correct = []
for index, row in match_df.iterrows():
    if row['Winner']==row['Batting Team Result DL']: DL_correct.append("Yes")
    else: DL_correct.append("No")

DL_correct_df = pd.DataFrame(DL_correct)
match_df['DL Correct?'] = DL_correct_df
```
