**SQL Analysis of Baseball stats from 1871 to 2015**

Data - https://www.kaggle.com/datasets/open-source-sports/baseball-databank/data?select=Fielding.csv

*Importing libraries*

```python
import pandas as pd
import sqlite3
```

*Importing datasets*

```python
master = pd.read_csv('Master.csv')
batting = pd.read_csv('Batting.csv')
fielding = pd.read_csv('Fielding.csv')
pitching = pd.read_csv('Pitching.csv')
```

*Connect to database*

```python
connection = sqlite3.connect('baseball.db')
master.to_sql('master', connection, if_exists='replace', index=False)
batting.to_sql('batting', connection, if_exists='replace', index=False)
fielding.to_sql('fielding', connection, if_exists='replace', index=False)
pitching.to_sql('pitching', connection, if_exists='replace', index=False)
```

*Explore the data*

```python
with sqlite3.connect('baseball.db') as connection:
    cursor = connection.cursor()
    query = 'SELECT * FROM  master'
    df = pd.read_sql_query(query, connection)
df
```

```python
df.isna().sum()
```

```python
cursor.execute("pragma table_info(master)")
columns_info = cursor.fetchall()
columns_info[0]
```

```python
col_info = pd.DataFrame(columns_info, columns=[d[0] for d in cursor.description])
col_info[['name', 'type']]
```

```python
cursor.close()
connection.close()
```
